```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './chipotle.csv'
```

```python
chipo=pd.read_csv(file_path)
```

```python
print(type(chipo))
```

```python
# 1. 음식 메뉴는 총 몇가지 일까?
# unique: 중복되는 데이터를 제거하는 함수
tmp_arr=chipo['item_name'].unique()
len(tmp_arr)
tmp_arr
```

```python
# 메뉴 당 가격은? 뭐가 제일 비싸지?
# item_price 칼럼의 문자열 $을 지우고 type을 실수로 표시함
chipo['item_price']=chipo['item_price'].str.lstrip('$')

chipo['item_price'].astype(float)

chipo
```

```python
chipo_cost=chipo.groupby('item_name')['item_price'].mean()
chipo_cost.sort_values(ascending=False)
```
