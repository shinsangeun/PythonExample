```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
file_path='./chipotle.csv'
```

```python
chipo=pd.read_csv(file_path)
chipo
```

```python
# 불필요한 컬럼 삭제
# inplace=True: 해당 변화를 즉각적으로 해당 DataFrame에 반영하겠다는 의미
chipo.drop(['order_id', 'item_price'], axis='columns', inplace=True)
chipo
```

```python
chipo['choice_description'].fillna('Origin',inplace=True)
chipo
```

```python
# 대괄호 지우기
num=0
for i in chipo['choice_description']:
    chipo['choice_description'][num]=re.sub(pattern='[^a-zA-Z,]',repl='',string=i)
    #[^a-zA-Z,]은 알파벳과 ',' , 공백을 제외하고 모두 지우겠다는 의미
    num+=1
chipo
```

```python
chipo['source']='new'
print(chipo['choice_description'][3])

```

```python
result_tmp=chipo.groupby(['item_name'])['choice_description'].value_counts()
result_tmp=result_tmp.to_frame()
result_tmp
```

```python
# 엑셀로 파일 저장
file_path="./chipotle_result.csv"
result_tmp.to_csv(file_path,sep=',')
```
